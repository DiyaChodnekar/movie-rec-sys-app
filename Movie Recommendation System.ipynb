# Defining Dataset

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head()
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

# Merging datasets

```python
movies.info()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'release_date', 'cast', 'crew']]
movies.head()
```

# Creation of Tags column
- Create a paragraph using specific words from genres, keywords, overview, cast, crew columns

# Data preprocessing
- Check for null values
- Extract required words/data to create tags
- Remove spaces between words (for e.g. X Y -> XY so that X A does not get included in X Y recommendations
- Stemming words (e.g. love, loves, loved reduce to love)

```python
# checking for null values
movies.isnull().sum()
```

```python
# dropping null values
movies.dropna(inplace=True)
```

```python
# checking duplicated values
movies.duplicated().sum()
```

## 1. genres

```python
# observation for first row
movies.iloc[0].genres
```

```python
# function to obtain list of column names from dict format
def convert(obj):
    list_ = []
    for i in ast.literal_eval(obj):
        list_.append(i['name'])
    return list_
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

## 2. keywords

```python
# checking first observation
movies.iloc[0].keywords
```

```python
# Applying above function to extract list of name key
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

## 3. cast

```python
# checking first observation
movies.iloc[0].cast
```

```python
# function to obtain list of first three actor names from dict format
def convert3(obj):
    list_ = []
    ctr = 0
    for i in ast.literal_eval(obj):
        if ctr != 3:
            list_.append(i['name'])
            ctr += 1
        else:
            break;
    return list_
```

```python
# Applying above function to extract list 
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

## 4. crew

```python
# checking first observation
movies.iloc[0].crew
```

```python
# function to obtain director name from dict format
def convert(obj):
    list_ = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list_.append(i['name'])
            break
    return list_
```

```python
# Applying above function to extract list 
movies['crew'] = movies['crew'].apply(convert)
movies.head()
```

## 5. overview

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

## Remove inter-word space

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

## Creating new dataframe with required columns
- Create a single tags column containing overview, genres, keywords, cast and crew

```python
# create new tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# creation of new dataframe
movies_df = movies[['movie_id', 'title', 'tags', 'release_date']]
movies_df.head()
```

```python
# convert tags column into string
movies_df['tags'] =  movies_df['tags'].apply(lambda x:" ".join(x))
movies_df.head()
```

```python
# convert contents of tags column to lowercase
movies_df['tags'] =  movies_df['tags'].apply(lambda x:x.lower())
movies_df.head()
```

## Stemming words in the tags column

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
# Helper function

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
movies_df.head()
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

# max_fetures: how many words to be considered
# stop_words: words like at, on, and ,are, etc
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
# vector of words contained in each movie
vectors = cv.fit_transform(movies_df['tags']).toarray()
print(vectors.shape)
vectors
```

```python
# View the top 5000 words
cv.get_feature_names()
```

# Computing cosine similarity
- Provides a number between 0 and 1, where 1 is most and 0 is least

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# shows the similarity of the first movie with all others
similarity[0]
```

```python
# Getting a list of similarities with their keys
list(enumerate(similarity[0]))
```

```python
# Sorting list in descending order based on similarities and getting the top 5 movies
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

# Recommend function

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
